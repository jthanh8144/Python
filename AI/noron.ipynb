```python
from sklearn import datasets
import math

iris = datasets.load_iris()
X = iris.data
Y = iris.target


def o(W, x):
    t = 0
    for i in range(len(x)):
        t += W[i]*x[i]
    ee = math.exp(t)
    return ee/(1+ee)


def e(W, X, y):
    sum = 0
    for i in range(len(X)):
        x = list(X[i])+[1]
        sum += (y[i] - o(W, x))**2
    return sum


def e1(W, i, X, y):
    d = 0.000001
    W1 = W[:]
    W1[i] = W1[i]+d
    return (e(W1, X, y)-e(W, X, y))/d


def perceptron(X, y):
    alpha = 1
    W = [0 for i in range(len(X[0])+1)]
    for _ in range(100):
        W_new = W[:]
        for i in range(len(W)):
            W_new[i] = W[i] - alpha * e1(W, i, X, y)
        e_new = e(W_new, X, y)
        e_ = e(W, X, y)
        if e_new > e_:
            alpha /= 2
        elif e_new < e_:
            W = W_new
            alpha = alpha * 2
        else:
            break
    return W


def model(j):
    y = list(Y)[:]
    for i in range(len(y)):
        if y[i] == j:
            y[i] = 1
        else:
            y[i] = 0
    return perceptron(X, y)


per = [model(i) for i in range(3)]

for i in range(len(X)):
    maxj = 0
    maxM = 0
    for j in range(len(per)):
        r = o(per[j], list(X[i])+[1])
        if maxM < r:
            maxM = r
            maxj = j
    print(Y[i], maxj)
```

```python
from sklearn import datasets
import math

iris = datasets.load_iris()
X = iris.data
Y = iris.target


def o(W, x):
    t = 0
    for i in range(len(x)):
        t += W[i]*x[i]
    ee = math.exp(t)
    return ee/(1+ee)


def O(WW, x):
    return [o(WW[i], x) for i in range(len(WW))]


def e(WW, X, Y):
    sum = 0
    for i in range(len(X)):
        x = list(X[i])+[1]
        t = O(WW, x)
        for j in range(len(t)):
            if Y[i] == j:
                sum += (1 - t[j])**2
            else:
                sum += t[j]**2
    return sum


def e1(WW, j, i, X, y):
    d = 0.000001
    W1 = [W[:] for W in WW]
    W1[j][i] = W1[j][i]+d
    return (e(W1, X, y)-e(WW, X, y))/d


def Model(X, Y):
    alpha = 1
    WW = [[0]*(len(X[0])+1)] * 3
    for _ in range(100):
        W_new = [W[:] for W in WW]
        for j in range(len(WW)):
            for i in range(len(WW[j])):
                W_new[j][i] = WW[j][i] - alpha * e1(WW, j, i, X, Y)
        e_new = e(W_new, X, Y)
        e_ = e(WW, X, Y)
        if e_new > e_:
            alpha /= 2
        elif e_new < e_:
            WW = W_new
            alpha = alpha * 2
        else:
            break
    return WW


per = Model(X, Y)

for i in range(len(X)):
    maxj = 0
    maxM = 0
    for j in range(len(per)):
        r = o(per[j], list(X[i])+[1])
        if maxM < r:
            maxM = r
            maxj = j
    print(Y[i], maxj)
```
